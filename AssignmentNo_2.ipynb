### Use Heuristic Search Techniques to Implement Best first search (Best-Solution but not always optimal) and A* algorithm (Always gives optimal solution).


```python
class Node:
    def __init__(self, f=0, g=999999, h=0, name=0):
        self.f = f
        self.g = g
        self.h = h
        self.name = name

    def setNeighbours(self, neighbours=[]):
        self.neighbours = neighbours

    def astar(self, goal, graph): # A STAR
        closedSet = set([])
        openSet = set([self])
        cameFrom = {}
        self.g = 0
        self.f = self.h

        while openSet:
            current = self.findNodeWithLowestFScore(openSet)
            if current == goal:
                return self.construct_path(cameFrom, current)

            openSet.remove(current)
            closedSet.add(current)

            for neighbour in current.neighbours:
                if neighbour in closedSet:
                    continue

                tentative_gScore = current.g + graph[current.name][neighbour.name]
                if tentative_gScore < neighbour.g:
                    cameFrom[neighbour] = current
                    neighbour.g = tentative_gScore
                    neighbour.f = neighbour.g + neighbour.h
                    if neighbour not in openSet:
                        openSet.add(neighbour)

        return -1

    def bfs(self, goal):  # BFS
        visited = set()
        queue = [self]
        cameFrom = {}

        while queue:
            current = queue.pop(0)
            if current == goal:
                return self.construct_path(cameFrom, current)

            visited.add(current)
            for neighbour in current.neighbours:
                if neighbour not in visited and neighbour not in queue:
                    cameFrom[neighbour] = current
                    queue.append(neighbour)

        return -1

    def findNodeWithLowestFScore(self, openSet):
        fScore = 999999
        node = None
        for eachNode in openSet:
            if eachNode.f < fScore:
                fScore = eachNode.f
                node = eachNode
        return node

    def construct_path(self, cameFrom, current):
        totalPath = []
        while current in cameFrom.keys():
            current = cameFrom[current]
            totalPath.append(current)
        return totalPath

graph = [
    [-1, 1, -1, -1, -1, 12],
    [1, -1, 2, 1, -1, -1],
    [-1, 2, -1, -1, 5, -1],
    [-1, 1, -1, -1, 3, 4],
    [-1, -1, 5, 3, -1, 2],
    [12, -1, -1, 4, 2, -1]
]

heuristics = [5, 3, 4, 2, 6, 0]
s = Node(h=heuristics[0], name=0)
a = Node(h=heuristics[1], name=1)
b = Node(h=heuristics[2], name=2)
c = Node(h=heuristics[3], name=3)
d = Node(h=heuristics[4], name=4)
g = Node(h=heuristics[5], name=5)

s.setNeighbours([a, g])
a.setNeighbours([c, b])
b.setNeighbours([a, d])
c.setNeighbours([a, d, g])
d.setNeighbours([b, c, g])
g.setNeighbours([s, c, d])

startNode = s
goalNode = g

if __name__ == "__main__":
    path = startNode.astar(goalNode, graph)
    print("A* Path is:", end=" ")
    for node in path[::-1]:
        print(str(node.name) + "-->", end="")
    print(goalNode.name)
    print("Cost =", goalNode.g)

    bfs_path = startNode.bfs(goalNode)
    print("\nBFS Path is:", end=" ")
    for node in bfs_path[::-1]:
        print(str(node.name) + "-->", end="")
    print(goalNode.name)
```
